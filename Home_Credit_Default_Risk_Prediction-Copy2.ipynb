# Data Preparation

```python
# Removing the ID of the person

Train_data = Train_data.drop("SK_ID_CURR",axis=1)
```

### Remove the High correlated features

```python
# Removing Highly Correlated featues

cor_matrix = Train_data.corr().abs()

upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))

to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.95)]

Train_data = Train_data.drop(Train_data[to_drop], axis=1)
```

### Removing the missing values features

```python
#Removing more than 60% missing values 

pct_null = Train_data.isnull().sum() / len(Train_data)

missing_features = pct_null[pct_null > 0.60].index

Train_data = Train_data.drop(missing_features, axis=1)
```

```python
# Here divinding target and features

Train_data1 = Train_data.drop("TARGET",axis=1)

y = Train_data[["TARGET"]]
```

```python
# Seperating the numerical and categorical features names for imputation

Numerical_Columns_Train_data = Train_data1.select_dtypes(include=["number"]).columns

Categorical_Columns_Train_data =Train_data1.select_dtypes(include=["object"]).columns
```

## Missing Values Imputation

```python
# Missing values imputation using SimpleImputer

from sklearn.impute import SimpleImputer

Simple_impute = SimpleImputer(strategy = 'median')

impute_Train_data = Simple_impute.fit_transform(Train_data1[Numerical_Columns_Train_data])

Train_data1 = Train_data1.drop(Train_data1[Numerical_Columns_Train_data],axis=1)

imputed_Train_data = pd.DataFrame(impute_Train_data,columns=Numerical_Columns_Train_data)

Train_data1 = pd.concat([Train_data1,imputed_Train_data],axis=1)
```

```python
#Missing values imputation for categorical featues

Train_data1[Categorical_Columns_Train_data] = Train_data1[Categorical_Columns_Train_data].fillna(Train_data1.mode().iloc[0])
```

```python
#Remaing after removing highly correlated featues and more than 60% missing values in the columns 

print("Shape of the data is :",Train_data1.shape)
```

**Cross validation is a technique for assessing how the statistical analysis generalises to an independent data set.It is a technique for evaluating machine learning models by training several models on subsets of the available input data and evaluating them on the complementary subset of the data.
